```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm, skew, kurtosis, shapiro, ks_2samp
from statsmodels.distributions.empirical_distribution import ECDF
from statsmodels.tsa.stattools import adfuller
```

# Download sp500 daily ohlc price for the last 365 days

```python
sp500_data = yf.download('^GSPC', period="1y")['Adj Close'].last('365D')
sp500_data
```

# Create variable for daily returns

```python
sp500_return = sp500_data.pct_change().dropna()
sp500_return
```

# Plot histogram

```python
sp500_return.hist(bins=20, color='red')
plt.title("histogram of sp500 return series")
plt.show()
```

# Calculate mean

```python
mean_ret = sp500_return.mean()
mean_ret
```

# Calculate variance

```python
var_ret = sp500_return.var() 
var_ret
```

# Calculate standard deviation

```python
sd_ret = sp500_return.std()
sd_ret
```

# Calculate kurtosis

```python
kurt_ret = kurtosis(sp500_return)
kurt_ret
```

# Sort returns from least to greatest

```python
sorted_returns = sp500_return.sort_values()
sorted_returns
```

# Calculate qth quantile for returns (i.e. q = 0.05)

```python
q_05 = sp500_return.quantile(0.05)
q_05
```

# Empirical cumulative distribution function

```python
f = ECDF(sp500_return)
plt.plot(f.x, f.y)
plt.show()
```

# P(-%2 < X < %1)

```python
diff = f(0.01) - f(-0.02)
diff
```

# Binomial distribution example for empirical VaR (Value-at-risk)

```python
days = range(len(sp500_return))
p_sum = [np.random.binomial(n, 0.05) for n in days]
p_sum = np.cumsum(p_sum)
plt.plot(p_sum)
plt.show()

# P(X = 4)
prob_x_4 = sum(1 for p in p_sum if p == 4) / len(p_sum)
prob_x_4
# P(X >= 10)
prob_x_10 = sum(1 for p in p_sum if p >= 10) / len(p_sum)
prob_x_10
```

# Fitting a normal distribution to data

```python
mu = mean_ret
st = sd_ret
x = np.linspace(mu - 4 * st, mu + 4 * st, 100)
plt.hist(sp500_return, bins=15, density=True)
plt.plot(x, norm.pdf(x, mu, st), color='red')
plt.show()
```

# QQ plot

```python
import statsmodels.api as sm
sm.qqplot(sp500_return, line ='45',fit=True)
```

# VaR under normal dist.

```python
qnorm = norm.ppf(0.05, mu, st)
qnorm
```

# Normalized returns

```python
zscore = (sp500_return - mean_ret) / sd_ret
import scipy.stats
_, bins, _ = plt.hist(zscore, 20, density=1, alpha=0.5)
mu, sigma = scipy.stats.norm.fit(zscore)
best_fit_line = scipy.stats.norm.pdf(bins, mu, sigma)
plt.plot(bins, best_fit_line)
```

# Download Microsoft and Apple historical prices

```python
msft_data = yf.download('MSFT', period="1y")['Adj Close'].last('365D')
aapl_data = yf.download('AAPL', period="1y")['Adj Close'].last('365D')
```

# Calculate returns

```python
Rm = msft_data.pct_change().dropna()
Ra = aapl_data.pct_change().dropna()

N = len(Rm)
```

```python
Rm , Ra
```

# Empirical joint probability that Ra < -0.02 and Rm < -0.02

```python
joint_prob = sum((Rm < -0.02) & (Ra < -0.02)) / N
joint_prob
```
